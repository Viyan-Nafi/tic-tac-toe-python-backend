```python

wins = [
    [1, 2, 3], [4, 5, 6], [7, 8, 9],  # horizontal wins
    [1, 4, 7], [2, 5, 8], [3, 6, 9],  # vertical wins
    [1, 5, 9], [3, 5, 7]              # diagonal wins
]

def print_board(step):
    print(f' {step[0]} | {step[1]} | {step[2]} ')
    print('-----------')
    print(f' {step[3]} | {step[4]} | {step[5]} ')
    print('-----------')
    print(f' {step[6]} | {step[7]} | {step[8]} ')

def tic_tac_toe():
    while True:
        # Initialize game variables at the start of each game
        step = [1, 2, 3, 4, 5, 6, 7, 8, 9]
        print('Welcome to TIC-TAC-TOE!')

        # Player choice
        p1 = ''
        while p1 not in ['X', 'O']:
            p1 = input('Player1: Do you want to be X or O? Please enter X or O: ').strip().upper()
            if p1 == 'X':
                p2 = 'O'
            else:
                p2 = 'X'
        print(f'Player1 is {p1}, Player2 is {p2}. Player1 goes first.')

        # Ready to play?
        start = ''
        while start not in ['yes', 'no']:
            start = input('Are you ready to play? Please enter yes or no: ').strip().lower()
        if start == 'no':
            print('Maybe next time. Goodbye!')
            break

        game_over = False
        current_player = p1
        move_count = 0

        # Main game loop
        while not game_over:
            print_board(step)
            pos_input = input('Choose your position (1-9): ').strip()

            # Validate input
            if not pos_input.isdigit() or int(pos_input) not in range(1,10):
                print('Invalid input! Please choose a number from 1 to 9.')
                continue

            pos = int(pos_input)
            # Check if position is already taken
            if isinstance(step[pos-1], str):
                print('Position already taken. Choose another one.')
                continue

            # Make move
            step[pos-1] = current_player
            move_count += 1

            # Check for a win
            for combo in wins:
                if step[combo[0]-1] == step[combo[1]-1] == step[combo[2]-1]:
                    print_board(step)
                    print(f'Congratulations! Player {current_player} wins!')
                    game_over = True
                    break

            # Check for a draw
            if move_count == 9 and not game_over:
                print_board(step)
                print('It\'s a draw!')
                game_over = True

            # Switch players
            if not game_over:
                if current_player == p1:
                    current_player = p2
                else:
                    current_player = p1

        # Ask to play again
        again = ''
        while again not in ['yes', 'no']:
            again = input('Do you want to play again? Enter yes or no: ').strip().lower()
        if again == 'no':
            print('Thanks for playing!')
            break

# Run the game
tic_tac_toe()
```
